```python
%matplotlib inline
# Dependencies and Setup 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data)
ride_data_df = pd.read_csv(ride_data)
# Combine the data into a single dataset
combined_data_df = pd.merge(ride_data_df, city_data_df, how='left', on='city')  
combined_data_df.head() 
# Display the data table for preview
```

```python
#Looking at DataFrame to see Driver_count
southm_df = combined_data_df[combined_data_df['city'] == 'South Michelleport']
southm_df
```

## Bubble Plot of Ride Sharing Data

```python
#Dataframe for each city type
rural_df =combined_data_df[combined_data_df['type'] == 'Rural']  
rural_df
urban_df =combined_data_df[combined_data_df['type'] == 'Urban'] 
urban_df
burb_df =combined_data_df[combined_data_df['type'] == 'Suburban'] 
burb_df                           
                          
```

```python
#Average Fare ($) Per City
#Series instead of new dataframe

rural_avg_fare= rural_df.groupby('city', as_index=False).mean()['fare'] 
rural_avg_fare

urban_avg_fare= urban_df.groupby('city', as_index=False).mean()['fare']
urban_avg_fare

burb_avg_fare= burb_df.groupby('city', as_index=False).mean()['fare']
burb_avg_fare




```

```python
#Total Number of Rides Per City
total_rural_rides = rural_df.groupby(["city"]).count()["ride_id"] 
total_rural_rides

total_urban_rides = urban_df.groupby(["city"]).count()["ride_id"]
total_urban_rides

total_burb_rides = burb_df.groupby(["city"]).count()["ride_id"] 
total_burb_rides
```

```python
#Total Number of Drivers Per City
rural_drivers_df = rural_df.groupby('city', as_index=False).mean()['driver_count'] 
rural_drivers_df

urban_drivers_df = urban_df.groupby('city', as_index=False).mean()['driver_count']
urban_drivers_df 

burb_drivers_df = burb_df.groupby('city', as_index=False).mean()['driver_count']
burb_drivers_df 



```

```python
#Obtain the x and y coordinates for each of the three city types (Urban, Suburban, Rural)
#Create scatter plot using a list

plt.scatter(total_rural_rides, rural_avg_fare, color='lightcoral', edgecolors="black", marker="o",  s=rural_drivers_df*10, alpha=0.75) 
plt.scatter(total_urban_rides, urban_avg_fare, color='lightskyblue', edgecolors="black", marker="o", s=urban_drivers_df*10, alpha=0.75)
plt.scatter(total_burb_rides, burb_avg_fare, color='gold', edgecolors="black", marker="o", s=burb_drivers_df*10, alpha=0.75)
plt.title('Pyber Ride Sharing Data (2016)', fontsize=14)
plt.xlabel('Total Number of Rides (Per City)', fontsize=14)
plt.ylabel('Average Fare ($)', fontsize=14)
plt.grid(True)
plt.gca().legend(('Rural','Urban', 'Suburban'))
plt.figtext(.95,.95,"Note: Circle size correlates with driver count per city.", wrap=True,
            verticalalignment='center_baseline', fontsize=12)

#Save Image 

plt.savefig('PyberBubbleChart.png') 
plt.show()
displayplot.png
```

## Total Fares by City Type

```python
#Calculate Type Percents
# Calculate percent of fares by city type
percent_fares = (combined_data_df.groupby(['type']).sum()['fare'] / (combined_data_df['fare'].sum())) * 100
percent_fares

# Build Pie Chart
#Labels for the sections of our pie chart
labels = ["Rural", "Suburbans", "Urban"]
#The values of each section of the pie chart
sizes = [6.8, 30.5, 62.7] 
#The colors of each section of the pie chart
colors = ['lightcoral', 'lightskyblue', 'gold']

#Title
plt.title("% of Total Fares by City Type")

#Tell matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

#Create pie chart based on values above
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=100)


plt.axis("equal")
# Save Figure
plt.savefig('TotalFaresbyCity.png') 
plt.show()
```
